```python
#Want to create a list of all ticker symbols in the S&P 500
#Import the requests package
import requests

#Define the URL for the S&P 500 list
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

#Get the URL
r = requests.get(url)

#Check the status code
r.status_code

#Get the content of the URL
html = r.content

#Import the BeautifulSoup package
from bs4 import BeautifulSoup

#Parse the HTML content
soup = BeautifulSoup(html, 'html.parser')

#Get the table with the S&P 500 list
table = soup.find_all('table')[0]

#Get the rows of the table
rows = table.find_all('tr')

#Create an empty list to store the ticker symbols
ticker_list = []

#Loop through the rows of the table
for row in rows[1:]:
    #Get the columns of the row
    cols = row.find_all('td')
    #Get the ticker symbol
    ticker = cols[0].text.strip()
    #Append the ticker symbol to the list
    ticker_list.append(ticker)

#Print the list of ticker symbols and its length
print(ticker_list, len(ticker_list))
```

```python
#use the Tickers module to get the data for all the stocks in the S&P 500
sp500_tickers = ' '.join(ticker_list)
sp500 = yf.Tickers(sp500_tickers)
```

### Below, we follow along with the youtube video on predicting Apple stock prices found [here](https://www.youtube.com/watch?v=bxC2Ilx5ErI).

```python
#import the usual packages for random-forest regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
stock = 'AAPL'
data = yf.download(stock, start="2020-01-01", end="2024-01-01") #download data
```

```python
#add previous day stock price
data['Previous Close'] = data['Close'].shift(1)

#add diff between high and low
data['High-Low'] = data['High'] - data['Low']

#add diff between open and close
data['Open-Close'] = data['Open'] - data['Close']

#add 10 and 50 day moving averages
data['MA_10_days'] = data['Close'].rolling(window=10).mean()
data['MA_50_days'] = data['Close'].rolling(window=50).mean()

#drop missing/NA values
data.dropna(inplace=True)
```

```python
#Define features
X = data[['Open', 'High', 'Low', 'Close', 'Volume', 'Previous Close', 'High-Low', 'Open-Close', 'MA_10_days', 'MA_50_days']]
y = data['Close']

#split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
#Create the model object
model = RandomForestRegressor(#random_state=42,
                              n_estimators=100)

#Fit the model
model.fit(X_train, y_train)

#Make predictions
y_pred = pd.Series(model.predict(X_test), index=X_test.index)

#Evaluate the model|
mse = MSE(y_test, y_pred)
print(mse)

#plot y_test and y_pred
plt.figure(figsize=(12, 10))
y_test.plot(linestyle='-', color='blue', label='Actual price')
y_pred.plot(linestyle='--', color='red', label='Predicted price')

plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```
